```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv("kenya_psa_dataset_translations.csv")

df.drop(columns=['en_len', 'sw_len','Target_Languages'], inplace=True)
```

```python
df.isnull().sum()
```

```python
#drop the kiswahili column
df = df.drop(columns=['Kiswahili'])
```

```python
df.head()
```

```python
print("Number of PSAs before dropping duplicates")
df.shape
```

```python
#drop duplicates based on the enlgish column
df2=df.drop_duplicates(subset=['English'])
print("Number of PSAs after dropping duplicates")
df2.shape
```

```python
# distribution of the English column by domain, create an interactive pie chart for this
domain_counts = df2['Domain'].value_counts().reset_index()
domain_counts.columns = ['Domain', 'Count']
import plotly.express as px
fig = px.pie(domain_counts, names='Domain', values='Count', title='Distribution of English PSAs by Domain', hole=0.4)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
# provide our top sources of PSAs per domain, create separate interactive bar charts for each domain
domain_groups = df2.groupby(['Domain', 'Source']).size().reset_index(name='Count')
for domain, group in domain_groups.groupby('Domain'):
    fig = px.bar(group.sort_values('Count', ascending=False),
                 x='Source',
                 y='Count',
                 color='Source',
                 color_discrete_sequence=px.colors.qualitative.Bold,
                 title=f'Top PSA Sources for {domain}',
                 labels={'Source': 'Source', 'Count': 'Number of PSAs'})
    fig.update_layout(xaxis_tickangle=-45,
                      margin=dict(l=40, r=40, t=80, b=140),
                      showlegend=False)
    fig.show()
```

```python
domain_groups
```

```python
# source distribution: interactive chart for source_counts
source_counts = df2['Source'].value_counts().reset_index()
source_counts.columns = ['Source', 'Count']
fig = px.bar(source_counts.sort_values('Count', ascending=False),
             x='Source',
             y='Count',
             color='Source',
             color_discrete_sequence=px.colors.qualitative.Vivid,
             title='Source Counts for English PSAs',
             labels={'Source': 'Source', 'Count': 'Number of PSAs'})
fig.update_layout(xaxis_tickangle=-45,
                  margin=dict(l=40, r=40, t=80, b=140),
                  showlegend=False)
fig.show()
```
